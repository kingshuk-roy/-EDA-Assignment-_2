```python
# Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

plt.figure(figsize=(8, 6))
plt.boxplot(df['Price'], vert=False)  # 'vert=False' to create a horizontal box plot
plt.title('Box Plot of Flight Prices')
plt.xlabel('Price')
plt.show()

'''
Certainly, I can help you with that. To identify potential outliers in the dataset, we need to visually inspect boxplots for relevant numerical features.

Without the actual data, I'll provide a general explanation of how to interpret boxplots and the potential impact of outliers on your analysis:

Let's assume we're analyzing the "Price" feature using a boxplot:

1. **Create a Boxplot:** A boxplot consists of a rectangular "box" that represents the interquartile range (IQR) of the data,
with a line inside the box indicating the median. The "whiskers" extend from the box to show the range within which most data points lie.
Any data points outside the whiskers are considered potential outliers.

2. **Identify Potential Outliers:** Outliers are data points that fall outside the whiskers of the boxplot.
They are significantly different from the rest of the data and can be either high or low values.

3. **Impact on Analysis:**

   - **Skewed Statistics:**
   Outliers can distort summary statistics such as the mean and standard deviation.
   The mean is sensitive to outliers and may not accurately represent the central tendency of the data.

   - **Misleading Visualizations:** Outliers can distort the scale of visualizations,
    making it challenging to observe meaningful patterns and trends among the majority of data points.

   - **Model Performance:** In predictive modeling, outliers can negatively impact model performance by introducing noise and leading to inaccurate predictions.

   - **Regression Analysis:** Outliers can influence regression coefficients, leading to misleading interpretations of the relationships between variables.

   - **Insights:** Outliers can lead to incorrect insights and conclusions, as they may represent unusual or rare events that don't reflect typical patterns.

   - **Decision-Making:** Decisions based on analyses that include outliers may not be well-founded or effective,
   as they may not reflect the true underlying behavior of the majority of cases.

   - **Data Integrity:** Outliers might also indicate data entry errors or anomalies in the data collection process, which need to be validated and addressed.

4. **Handling Outliers:**

   - **Investigate:** First, verify whether the potential outliers are valid data points or errors. Domain knowledge is crucial for making this determination.

   - **Contextual Analysis:** If the outliers are legitimate, consider analyzing them separately or understanding the context in which they occur.

   - **Transformation:** Depending on the distribution, you might consider data transformations to reduce the impact of outliers on your analysis.

   - **Model Selection:** Choose models that are robust to outliers or use techniques like robust regression.

   - **Reporting:** Clearly report the presence of outliers in your analysis and explain their potential influence on your results and conclusions.

Remember that outliers should be carefully considered based on the specific context of your analysis.
They can provide valuable insights into the data distribution and underlying patterns,
but they can also introduce bias and lead to incorrect conclusions if not properly handled.'''
```

```python
# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

# Clean 'Size' column and convert it to numeric
df1['Size'] = df1['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
df1['Size'] = df1['Size'].apply(lambda x: str(x).replace('k', '') if 'k' in str(x) else x)
df1['Size'] = pd.to_numeric(df1['Size'], errors='coerce')

# Drop rows with missing values in 'Size' or 'Rating'
df1.dropna(subset=['Size', 'Rating'], inplace=True)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Size', y='Rating', data=df1, alpha=0.8)
plt.title('App Size vs. Rating')
plt.xlabel('App Size (MB)')
plt.ylabel('App Rating')
plt.grid(True)
plt.show()
```

```python
# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

df1["Price"].unique()
```

```python
chars_to_remove=["+",",","$"]
cols_to_clean=['Installs','Price']
for i in chars_to_remove:
    for j in cols_to_clean:
      df1[j]=df1[j].str.replace(i,'')
df1["Installs"]=df1["Installs"].astype(int)
df1["Price"]=df1["Price"].astype(float)
```

```python
df1.info()
```

```python
# Calculate average prices by app type
average_prices = df1.groupby('Type')['Price'].mean()

# Create the bar chart
plt.figure(figsize=(8, 6))
average_prices.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.tight_layout()
```
